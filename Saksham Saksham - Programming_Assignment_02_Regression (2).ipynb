```python
import numpy as np
import matplotlib.pyplot as plt

########################################
#Generate meaningfull data
########################################
def equation(x):
    return 3.86 * x - 110.42

num_points = 25
x_values = np.linspace(60, 80, num_points)
true_y_values = equation(x_values)

mean = 0
variance = 20
noise = np.random.normal(mean, np.sqrt(variance), num_points)
y_with_noise = true_y_values + noise
########################################
#Plot scatter plot of data
########################################
plt.figure(figsize=(10, 6))
plt.scatter(x_values, y_with_noise, color='green', label='Data Points')
plt.xlabel("Human Height (in inches)")
plt.ylabel("Weight (in pounds)")
plt.legend()
plt.title("Scatter Plot of Data")
plt.show()
########################################
#Weight estimation through error surface, i.e., empirically locate the minima of error surface
########################################
#Sample a bunch of w's around w_opt and compute the associated error
w0_values = np.linspace(-200, 0, 100)
w1_values = np.linspace(2, 6, 100)
W0, W1 = np.meshgrid(w0_values, w1_values)
#Compute the error
def Error(w, t, x):
    estimate = w[0] + w[1] * x
    error = 0.5 * np.sum((estimate - t) ** 2)
    return error

error_values = np.zeros_like(W0)

for i in range(len(w0_values)):
    w0 = w0_values[i]
    for j in range(len(w1_values)):
        w1 = w1_values[j]
        error = Error([w0, w1], y_with_noise, x_values)
        error_values[i, j] = error


#Plot 3D error surface and the corresponding contour plots
#Error surface is a function of w0 and w1
# Plot 3D error surface
fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(1, 2, 1, projection='3d')
surf = ax.plot_surface(W0, W1, error_values, cmap='viridis')
ax.set_xlabel('w0')
ax.set_ylabel('w1')
ax.set_zlabel('Error')
ax.set_title('3D Error Surface')


ax = fig.add_subplot(1, 2, 2)
plt.contour(W0, W1, error_values, levels=20, cmap='viridis')
plt.xlabel('w0')
plt.ylabel('w1')
plt.title('Contour Plot of Error Surface')

# Locate the minima of the error surface
minima = np.unravel_index(np.argmin(error_values), error_values.shape)
minimum_wo = W0[minima]
minimum_w1 = W1[minima]
minimum_error = error_values[minima]
ax.scatter(minimum_wo, minimum_w1, color='red', s=100, label=f'Minimum Error: {minimum_error:.2f}')
ax.legend()
plt.show()

########################################
#Least squares approach to estimate the weights
########################################
#Complete the below linear regression function
def LinearRegression(x, t):
    X = np.column_stack((np.ones_like(x), x))
    w_opt = np.linalg.lstsq(X, t, rcond=None)[0]
    return w_opt

#Estimate optimal weight's using "LinearRegression" function
w_ls = LinearRegression(x_values, y_with_noise)
w0_ls, w1_ls = w_ls
targets_estimated = equation(x_values)

#Plot the estimated line on top of the above scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x_values, y_with_noise, color='green', label='Data Points')
plt.xlabel("Human Height (in inches)")
plt.ylabel("Weight (in pounds)")
plt.plot(x_values, targets_estimated, color='blue', label='Estimated Line')
plt.legend()
plt.title("Estimated Line using Least Squares Approach")
plt.show()

########################################
#Compare the estimated weight's using least squares approach with the error surface approach
########################################
print("Estimated weights using least squares approach:")
print("w0_ls:", w0_ls)
print("w1_ls:", w1_ls)
print("\nEstimated weights using error surface approach:")
print("Minimum Error:", minimum_error)
print("w0 from error surface approach:", minimum_wo)
print("w1 from error surface approach:", minimum_w1)
```

```python
#Understanding model order and overfitting
########################################
#Generate 20 data points
########################################
num_points_2 = 20
e = np.random.normal(0, 0.1, num_points_2)
def function(x, e):
    return np.sin(2 * np.pi * x) + e
x_values2 = np.random.uniform(0, 1, size = 20)
y_values = function(x_values2, e)

#Obtain train and test splits
train_data_x = x_values2[:10]
train_data_y = y_values[:10]

test_data_x = x_values2[10:]
test_data_y = y_values[10:]


########################################
#Fitting Mth degree polynomial using least squares approach
########################################
#Complete the function
def PolynomialFit(X_train,Y_train,M,lamda = 0): #(training data, trining targets, Model order, Regularization coefficient)
    #Transform the data using polynomial kernel
    transformed_features = np.vander(X_train, M + 1, increasing=True)
    #Find Pseudo inverse solution
    transformed_transpose = transformed_features.T
    result = np.dot(transformed_transpose, transformed_features)
    regularisation = lamda * np.eye(*result.shape)
    sum = result + regularisation
    pseudo_inverse = np.linalg.pinv(sum)
    result2 = np.dot(pseudo_inverse, transformed_transpose)
    w_opt = np.dot(result2, Y_train)
    #return the weight vector
    return w_opt

#Complete the function
def PolynomialPred(w_est,X_train,X_test): #(weight,training data, testing data, training targets, testing targets)
    #Estimate the targets for both training and testing data
    M = len(w_est)
    transformed_features_train = np.vander(X_train, M , increasing=True)
    transformed_features_test = np.vander(X_test, M , increasing=True)

    trainPred = np.dot(transformed_features_train, w_est)
    testPred = np.dot(transformed_features_test, w_est)

    #Return training and testing predictions
    return trainPred, testPred

#Complete the function
def PolynomialPred_Error(w_est,X_train,X_test,Y_train,Y_test): #(weight,training data, testing data, training targets, testing targets)
    #Estimate the targets for both training and testing data
    M = len(w_est)
    transformed_features_train = np.vander(X_train, M , increasing=True)
    transformed_features_test = np.vander(X_test, M , increasing=True)

    trainPred = np.dot(transformed_features_train, w_est)
    testPred = np.dot(transformed_features_test, w_est)

    trainError = np.sqrt(np.sum((trainPred - Y_train) ** 2))
    testError = np.sqrt(np.sum((testPred - Y_test) ** 2))

    #Return training and testing error
    return trainError,testError

#Iterate through range of M values
M_range=list(range(10))
TrError = []
TeError = []
Y_Pred = [[], []]
for M in M_range:
    #Fit Mth order polynomial i.e estimate optimal w. Use the function "PolynomialFit"
    w_est_20 = PolynomialFit(train_data_x, train_data_y, M)
    #Predict training and testing targets
    TrainPred, TestPred = PolynomialPred(w_est_20, train_data_x, test_data_x)
    Y_Pred[0].append(TrainPred)
    Y_Pred[1].append(TestPred)
    #Predict errors on both training and testing data using estimated w. Use the function "PolynomialPred_Error"
    TrainError, TestError = PolynomialPred_Error(w_est_20, train_data_x, test_data_x, train_data_y, test_data_y)

    #Store them for plotting
    TrError.append(TrainError)
    TeError.append(TestError)

#Plot training and testing estimates alogwith the original targets
for i in range(len(Y_Pred[0])):
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))

    for ax in axs.flat:
        ax.set_xlabel('X')
        ax.set_ylabel('Y')

    axs[0].scatter(train_data_x, train_data_y, label='Train Data')
    axs[0].scatter(train_data_x, Y_Pred[0][i], label=f'Predictions on Train Data (Polynomial) - Degree {i}')
    axs[0].legend(loc='upper left')

    axs[1].scatter(test_data_x, test_data_y, label='Test Data')
    axs[1].scatter(test_data_x, Y_Pred[1][i], label=f'Predictions on Test Data (Polynomial) - Degree {i}')
    axs[1].legend(loc='upper left')

    axs[0].set_title(f'Train Data - Degree {i+1}')
    axs[1].set_title(f'Test Data - Degree {i+1}')

    plt.tight_layout()
    plt.show()


#Plot training error vs polynomial order, and testing error vs polynomial order
plt.plot(M_range, TrError, label = "Training Error")
plt.plot(M_range, TeError, label = "Testing Error")
plt.xlabel("Polynomial Order")
plt.ylabel("Training Error")
plt.title("Training and Testing Error Vs Polynomial Order")
plt.legend()
plt.show()


########################################
#Increase the size of training data set to 100 points and repeat the experiments
########################################
num_points_3 = 100
e2 = np.random.normal(0, 0.1, num_points_3)
x_values_100 = np.random.uniform(0, 1, size = 100)
y_values_100 = function(x_values_100, e2)

train_data_x_100 = x_values_100[:50]
train_data_y_100 = y_values_100[:50]

test_data_x_100 = x_values_100[50:]
test_data_y_100 = y_values_100[50:]

TrError_100 = []
TeError_100 = []
Y_Pred_100 = [[], []]

for M in M_range:
    w_est_100 = PolynomialFit(train_data_x_100, train_data_y_100, M)
    # Predict training and testing targets
    train_pred_100, test_pred_100 = PolynomialPred(w_est_100, train_data_x_100, test_data_x_100)
    Y_Pred_100[0].append(train_pred_100)
    Y_Pred_100[1].append(test_pred_100)
    # Predict errors on both training and testing data using estimated w
    train_error_100, test_error_100 = PolynomialPred_Error(w_est_100, train_data_x_100, test_data_x_100, train_data_y_100, test_data_y_100)
    TrError_100.append(train_error_100)
    TeError_100.append(test_error_100)

for i in range(len(Y_Pred_100[0])):
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))

    for ax in axs.flat:
        ax.set_xlabel('X')
        ax.set_ylabel('Y')

    axs[0].scatter(train_data_x_100, train_data_y_100, label='Train Data')
    axs[0].scatter(train_data_x_100, Y_Pred_100[0][i], label=f'Predictions on Train Data (Polynomial) - Degree {i}')
    axs[0].legend(loc='upper left')

    axs[1].scatter(test_data_x_100, test_data_y_100, label='Test Data')
    axs[1].scatter(test_data_x_100, Y_Pred_100[1][i], label=f'Predictions on Test Data (Polynomial) - Degree {i}')
    axs[1].legend(loc='upper left')

    axs[0].set_title(f'Train Data(100 Points) - Degree {i+1}')
    axs[1].set_title(f'Test Data(100 Points) - Degree {i+1}')

    plt.tight_layout()
    plt.show()


#Plot training error vs polynomial order, and testing error vs polynomial order
plt.plot(M_range, TrError_100, label = "Training Error")
plt.plot(M_range, TeError_100, label = "Testing Error")
plt.xlabel("Polynomial Order")
plt.ylabel("Training Error")
plt.title("Training and Testin Error Vs Polynomial Order (100 points)")
plt.legend()
plt.show()
########################################
#Effect of regularization
########################################
#Consider a set of lambda's. For example: lamdas = [0, 1e-7 , 1e-4, 1e-2, 1]
#Repeat the experiments, i.e., plot the prediction and error in predictions with respect to model order. Contrast these results with those obtained without regularization.
lamdas = [0, 1e-7, 1e-4, 1e-2, 1]
TrError_lam_20 = []
TeError_lam_20 = []
Y_Pred_lam_20 = [[], []]

for M in M_range:
    TrError_lam_inner = []
    TeError_lam_inner = []
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))

    for lamda in lamdas:
        w_est_lam = PolynomialFit(train_data_x, train_data_y, M, lamda)
        train_pred_lam_20, test_pred_lam_20 = PolynomialPred(w_est_lam, train_data_x, test_data_x)
        Y_Pred_lam_20[0].append(train_pred_lam_20)
        Y_Pred_lam_20[1].append(test_pred_lam_20)
        train_error_lam_20, test_error_lam_20 = PolynomialPred_Error(w_est_lam, train_data_x, test_data_x, train_data_y, test_data_y)
        TrError_lam_inner.append(train_error_lam_20)
        TeError_lam_inner.append(test_error_lam_20)

        axs[0].scatter(train_data_x, train_data_y, label='Train Data', color='blue')
        axs[0].scatter(train_data_x, Y_Pred_lam_20[0][-1], label=f'Predictions- Lambda={lamda}', alpha=0.5)
        axs[0].legend(loc='upper right')

        axs[1].scatter(test_data_x, test_data_y, label='Test Data', color='green')
        axs[1].scatter(test_data_x, Y_Pred_lam_20[1][-1], label=f'Predictions - Lambda={lamda}', alpha=0.5)
        axs[1].legend(loc='upper right')

        axs[0].set_title(f'Train Data(20 Points) - Degree {M}')
        axs[1].set_title(f'Test Data(20 Points) - Degree {M}')

    plt.tight_layout()
    plt.show()

    TrError_lam_20.append(TrError_lam_inner)
    TeError_lam_20.append(TeError_lam_inner)



TrError_lam_plot = []
TeError_lam_plot = []

for lamda in lamdas:
    TrError_lam_inner_100 = []
    TeError_lam_inner_100 = []
    for M in M_range:
        w_est_lam_100 = PolynomialFit(train_data_x_100, train_data_y_100, M, lamda)
        train_error_lam_100, test_error_lam_100 = PolynomialPred_Error(w_est_lam_100, train_data_x_100, test_data_x_100, train_data_y_100, test_data_y_100)
        TrError_lam_inner_100.append(train_error_lam_100)
        TeError_lam_inner_100.append(test_error_lam_100)
    TrError_lam_plot.append(TrError_lam_inner_100)
    TeError_lam_plot.append(TeError_lam_inner_100)
for i, lamda in enumerate(lamdas):
    plt.plot(M_range, TrError_lam_plot[i], label=f"λ={lamda}")
    # plt.plot(M_range, TeError_lam_plot[i], label=f"Testing Error (λ={lamda})")
plt.xlabel("Polynomial Order")
plt.ylabel("Error")
plt.title("Effect of Regularization on Training Data(20 Points)")
plt.legend()
plt.show()
for i, lamda in enumerate(lamdas):
    # plt.plot(M_range, TrError_lam_plot[i], label=f"Training Error (λ={lamda})")
    plt.plot(M_range, TeError_lam_plot[i], label=f"λ={lamda}")
plt.xlabel("Polynomial Order")
plt.ylabel("Error")
plt.title("Effect of Regularization on Testing Data(20 Points)")
plt.legend()
plt.show()


lamdas = [0, 1e-7, 1e-4, 1e-2, 1]
TrError_lam_20 = []
TeError_lam_20 = []
Y_Pred_lam_20 = [[], []]

for M in M_range:
    TrError_lam_inner = []
    TeError_lam_inner = []
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))

    for lamda in lamdas:
        w_est_lam = PolynomialFit(train_data_x_100, train_data_y_100, M, lamda)
        train_pred_lam_20, test_pred_lam_20 = PolynomialPred(w_est_lam, train_data_x_100, test_data_x_100)
        Y_Pred_lam_20[0].append(train_pred_lam_20)
        Y_Pred_lam_20[1].append(test_pred_lam_20)
        train_error_lam_20, test_error_lam_20 = PolynomialPred_Error(w_est_lam, train_data_x_100, test_data_x_100, train_data_y_100, test_data_y_100)
        TrError_lam_inner.append(train_error_lam_20)
        TeError_lam_inner.append(test_error_lam_20)

        axs[0].scatter(train_data_x_100, train_data_y_100, label='Train Data', color='blue')
        axs[0].scatter(train_data_x_100, Y_Pred_lam_20[0][-1], label=f'Predictions- Lambda={lamda}', alpha=0.5)
        axs[0].legend(loc='upper right')

        axs[1].scatter(test_data_x_100, test_data_y_100, label='Test Data', color='green')
        axs[1].scatter(test_data_x_100, Y_Pred_lam_20[1][-1], label=f'Predictions - Lambda={lamda}', alpha=0.5)
        axs[1].legend(loc='upper right')

        axs[0].set_title(f'Train Data(100 Points) - Degree {M}')
        axs[1].set_title(f'Test Data(100 Points) - Degree {M}')

    plt.tight_layout()
    plt.show()

    TrError_lam_20.append(TrError_lam_inner)
    TeError_lam_20.append(TeError_lam_inner)



TrError_lam_plot = []
TeError_lam_plot = []

for lamda in lamdas:
    TrError_lam_inner_100 = []
    TeError_lam_inner_100 = []
    for M in M_range:
        w_est_lam_100 = PolynomialFit(train_data_x_100, train_data_y_100, M, lamda)
        train_error_lam_100, test_error_lam_100 = PolynomialPred_Error(w_est_lam_100, train_data_x_100, test_data_x_100, train_data_y_100, test_data_y_100)
        TrError_lam_inner_100.append(train_error_lam_100)
        TeError_lam_inner_100.append(test_error_lam_100)
    TrError_lam_plot.append(TrError_lam_inner_100)
    TeError_lam_plot.append(TeError_lam_inner_100)
for i, lamda in enumerate(lamdas):
    plt.plot(M_range, TrError_lam_plot[i], label=f"λ={lamda}")
    # plt.plot(M_range, TeError_lam_plot[i], label=f"Testing Error (λ={lamda})")
plt.xlabel("Polynomial Order")
plt.ylabel("Error")
plt.title("Effect of Regularization on Training Data(100 Points)")
plt.legend()
plt.show()
for i, lamda in enumerate(lamdas):
    # plt.plot(M_range, TrError_lam_plot[i], label=f"Training Error (λ={lamda})")
    plt.plot(M_range, TeError_lam_plot[i], label=f"λ={lamda}")
plt.xlabel("Polynomial Order")
plt.ylabel("Error")
plt.title("Effect of Regularization on Testing Data(100 Points)")
plt.legend()
plt.show()




########################################
#Effect of bias regularization
########################################
#Modify the function i.e include bias
#Generate data
num_points_bias = 100
e = np.random.normal(0, 0.1, num_points_bias)
def bias_equation(x, e, bias):
    return bias + np.sin(2 * np.pi * x) + e

x_values_bias_20 = np.random.uniform(0, 1, size = 100)
bias = 5
bias_values = [5]
for bias in bias_values:
    y_values_bias_20 = bias_equation(x_values_bias_20, e, bias)


    train_data_x_100 = x_values_bias_20[:50]
    train_data_y_100 = y_values_bias_20[:50]

    test_data_x_100 = x_values_bias_20[50:]
    test_data_y_100 = y_values_bias_20[50:]

lamdas = [0, 1e-7, 1e-4, 1e-2, 1]
TrError_lam_20 = []
TeError_lam_20 = []
Y_Pred_lam_20 = [[], []]

for M in M_range:
    TrError_lam_inner = []
    TeError_lam_inner = []
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))

    for lamda in lamdas:
        w_est_lam = PolynomialFit(train_data_x_100, train_data_y_100, M, lamda)
        train_pred_lam_20, test_pred_lam_20 = PolynomialPred(w_est_lam, train_data_x_100, test_data_x_100)
        Y_Pred_lam_20[0].append(train_pred_lam_20)
        Y_Pred_lam_20[1].append(test_pred_lam_20)
        train_error_lam_20, test_error_lam_20 = PolynomialPred_Error(w_est_lam, train_data_x_100, test_data_x_100, train_data_y_100, test_data_y_100)
        TrError_lam_inner.append(train_error_lam_20)
        TeError_lam_inner.append(test_error_lam_20)

        axs[0].scatter(train_data_x_100, train_data_y_100, label='Train Data', color='blue')
        axs[0].scatter(train_data_x_100, Y_Pred_lam_20[0][-1], label=f'Predictions- Lambda={lamda}', alpha=0.5)
        axs[0].legend(loc='upper right')

        axs[1].scatter(test_data_x_100, test_data_y_100, label='Test Data', color='green')
        axs[1].scatter(test_data_x_100, Y_Pred_lam_20[1][-1], label=f'Predictions - Lambda={lamda}', alpha=0.5)
        axs[1].legend(loc='upper right')

        axs[0].set_title(f'Train Data(Bias = {bias} - Degree {M})')
        axs[1].set_title(f'Test Data(Bias = {bias} - Degree {M})')

    plt.tight_layout()
    plt.show()

    TrError_lam_20.append(TrError_lam_inner)
    TeError_lam_20.append(TeError_lam_inner)



TrError_lam_plot = []
TeError_lam_plot = []

for lamda in lamdas:
    TrError_lam_inner_100 = []
    TeError_lam_inner_100 = []
    for M in M_range:
        w_est_lam_100 = PolynomialFit(train_data_x_100, train_data_y_100, M, lamda)
        train_error_lam_100, test_error_lam_100 = PolynomialPred_Error(w_est_lam_100, train_data_x_100, test_data_x_100, train_data_y_100, test_data_y_100)
        TrError_lam_inner_100.append(train_error_lam_100)
        TeError_lam_inner_100.append(test_error_lam_100)
    TrError_lam_plot.append(TrError_lam_inner_100)
    TeError_lam_plot.append(TeError_lam_inner_100)
for i, lamda in enumerate(lamdas):
    plt.plot(M_range, TrError_lam_plot[i], label=f"λ={lamda}")
    # plt.plot(M_range, TeError_lam_plot[i], label=f"Testing Error (λ={lamda})")
plt.xlabel("Polynomial Order")
plt.ylabel("Error")
plt.title(f"Effect of Regularization on Training Data with Bias = {bias}")
plt.legend()
plt.show()
for i, lamda in enumerate(lamdas):
    # plt.plot(M_range, TrError_lam_plot[i], label=f"Training Error (λ={lamda})")
    plt.plot(M_range, TeError_lam_plot[i], label=f"λ={lamda}")
plt.xlabel("Polynomial Order")
plt.ylabel("Error")
plt.title(f"Effect of Regularization on Testing Data with bias = {bias}")
plt.legend()
plt.show()



#Estimate the polynomial with and without regularization constraint



#Compare the two estimated polynomials and report the observations

```

```python
#Understanding the choice of kernel
########################################
#Generate 100 data points
########################################
num_points_p3 = 100
e = np.random.normal(0, 0.1, num_points_p3)
def function(x, e):
    return np.sin(2 * np.pi * x) + e
x_values_p3 = np.random.uniform(0, 1, size = 100)
y_values_p3 = function(x_values_p3, e)

train_data_x_p3 = x_values_p3[:50]
train_data_y_p3 = y_values_p3[:50]

test_data_x_p3 = x_values_p3[50:]
test_data_y_p3 = y_values_p3[50:]


#Obtian train and test splits
#Take even samples for training and odd samples for testing



#Function to estimate the parameters
def KernelRegressionFit(X_train,Y_train,kernelType,M,lamda): #(training data, training targets, type of kernel, regularization coefficient)
    #kernelType : {Polynomial,Gaussian,Sigmoid}
    #print(X_train.shape)
    #X_train = np.reshape(X_train, (1,-1))
    #Y_train = np.reshape(Y_train, (1,-1))
  # X = []
  if kernelType=='polynomial':
      #Use polynomial kernel to transform the data
    transformed_features = np.vander(X_train, M + 1, increasing=True)
  if kernelType=='gaussian':
    #Use Gaussian kernel to transform the data

    sigma = 1
    n_samples = len(X_train)
    transformed_features = np.zeros((n_samples, M + 1))
    intervals = np.linspace(min(X_train), max(X_train), M + 1)
    for i, interval in enumerate(intervals):
        transformed_features[:, i] = np.exp(-((X_train - interval) ** 2) / (2 * sigma ** 2))

  if kernelType=='sigmoidal':
      #Use Sigmoid kernel to transform the data
    gamma = 1.0
    coeff = 0.1
    n_samples = len(X_train)
    transformed_features = np.zeros((n_samples, M + 1))
    intervals = np.linspace(min(X_train), max(X_train), M + 1)
    for i, interval in enumerate(intervals):
        transformed_features[:, i] = np.tanh(gamma * (X_train - interval) + coeff)


  #Estimate weights using Pseudo iverse solution
  transformed_transpose = transformed_features.T
  result = np.dot(transformed_transpose, transformed_features)
  # print(transformed_features.shape)  # Inspect shape of transformed features
  # print(result.shape)  # Check the shape of the 'result' matrix
  regularisation = lamda * np.eye(*result.shape)
  sum = result + regularisation
  pseudo_inverse = np.linalg.inv(sum)
  result2 = np.dot(pseudo_inverse, transformed_transpose)
  w_opt = np.dot(result2, Y_train)

    #Return the estimated weights
  return w_opt

#Function to compute the training and testing errors from the current weight estimates
def KernelRegressionPred_Error(w_est,X_train,Y_train,X_test,Y_test,kernelType):
  #kernelType : {Polynomial,Gaussian,Sigmoid}
  X_tr = []
  X_te = []
  #X_train = np.reshape(X_train, (1,-1))
  #Y_train = np.reshape(Y_train, (1,-1))
  #X_test = np.reshape(X_test, (1,-1))
  #Y_test = np.reshape(Y_test, (1,-1))
  M = len(w_est)-1
  if kernelType=='polynomial':
    #Use polynomial kernel to transform the data
    transformed_features_train = np.vander(X_train, M + 1, increasing=True)
    transformed_features_test = np.vander(X_test, M + 1, increasing=True)


  if kernelType=='gaussian':
    #Use Gaussian kernel to transform the data
    sigma = 1
    n_samples_train = len(X_train)
    transformed_features_train = np.zeros((n_samples_train, M + 1))
    intervals = np.linspace(min(X_train), max(X_train), M + 1)
    for i, interval in enumerate(intervals):
        transformed_features_train[:, i] = np.exp(-((X_train - interval) ** 2) / (2 * sigma ** 2))

    n_samples_test = len(X_test)
    transformed_features_test = np.zeros((n_samples_test, M + 1))
    intervals_test = np.linspace(min(X_test), max(X_test), M + 1)
    for i, interval in enumerate(intervals_test):
        transformed_features_test[:, i] = np.exp(-((X_test - interval) ** 2) / (2 * sigma ** 2))

  if kernelType=='sigmoidal':
    gamma = 1.0
    coeff = 0.1
    n_samples_train = len(X_train)
    transformed_features_train = np.zeros((n_samples_train, M + 1))
    intervals = np.linspace(min(X_train), max(X_train), M + 1)
    for i, interval in enumerate(intervals):
        transformed_features_train[:, i] = np.tanh(gamma * (X_train - interval) + coeff)

    n_samples_test = len(X_test)
    transformed_features_test = np.zeros((n_samples_test, M + 1))
    intervals_test = np.linspace(min(X_test), max(X_test), M + 1)
    for i, interval in enumerate(intervals_test):
        transformed_features_test[:, i] = np.tanh(gamma * (X_test - interval) + coeff)

  #Estimate training and testing targets
  trainPred = np.dot(transformed_features_train, w_est)
  testPred = np.dot(transformed_features_test, w_est)

  trainError = np.sqrt(np.sum((trainPred - Y_train) ** 2))
  testError = np.sqrt(np.sum((testPred - Y_test) ** 2))


  #Compute and return the training and testing errors
  return trainPred, testPred, trainError,testError


#Iterate through range of M values
M_range=list(range(10))

polynomial_tr_error = []
polynomial_te_error = []
gaussian_tr_error = []
gaussian_te_error = []
sigmoidal_tr_error = []
sigmoidal_te_error = []
Y_Pred_polynomial = [[],[]]
Y_Pred_gaussian = [[],[]]
Y_Pred_sigmoidal = [[],[]]

for M in M_range:
    #Fit Mth order polynomial using three kernels i.e {Polynomial,Gaussian,Sigmoid}
  w_est_polynomial = KernelRegressionFit(train_data_x_p3, train_data_y_p3, 'polynomial', M, 0)
  w_est_gaussian = KernelRegressionFit(train_data_x_p3, train_data_y_p3, 'gaussian', M, 0)
  w_est_sigmoidal = KernelRegressionFit(train_data_x_p3, train_data_y_p3, 'sigmoidal', M, 0)

  trainpred_polynomial, testpred_polynomial, trainerror_polynomial, testerror_polynomial = KernelRegressionPred_Error(w_est_polynomial, train_data_x_p3, train_data_y_p3, test_data_x_p3, test_data_y_p3, 'polynomial')
  Y_Pred_polynomial[0].append(trainpred_polynomial)
  Y_Pred_polynomial[1].append(testpred_polynomial)
  polynomial_tr_error.append(trainerror_polynomial)
  polynomial_te_error.append(testerror_polynomial)


  trainpred_gaussian, testpred_gaussian, trainerror_gaussian, testerror_gaussian = KernelRegressionPred_Error(w_est_gaussian, train_data_x_p3, train_data_y_p3, test_data_x_p3, test_data_y_p3, 'gaussian')
  Y_Pred_gaussian[0].append(trainpred_gaussian)
  Y_Pred_gaussian[1].append(testpred_gaussian)
  gaussian_tr_error.append(trainerror_gaussian)
  gaussian_te_error.append(testerror_gaussian)

  trainpred_sigmoidal, testpred_sigmoidal, trainerror_sigmoidal, testerror_sigmoidal = KernelRegressionPred_Error(w_est_sigmoidal, train_data_x_p3, train_data_y_p3, test_data_x_p3, test_data_y_p3, 'sigmoidal')
  Y_Pred_sigmoidal[0].append(trainpred_sigmoidal)
  Y_Pred_sigmoidal[1].append(testpred_sigmoidal)
  sigmoidal_tr_error.append(trainerror_sigmoidal)
  sigmoidal_te_error.append(testerror_sigmoidal)



    #Predict training and testing targets using estimated w


    #Predict errors on both training and testing data using estimated w



    #Store them for plotting



for i in range(len(Y_Pred_polynomial[0])):
  fig, axs = plt.subplots(2, 3, figsize=(18, 12))
  fig.suptitle('Comparison of Predictions for Different Kernels and Polynomial Degrees', fontsize=16)
  for ax in axs.flat:
      ax.set_xlabel('X')
      ax.set_ylabel('Y')


  axs[0, 0].scatter(train_data_x_p3, train_data_y_p3, label='Train Data')
  axs[0, 0].scatter(train_data_x_p3, Y_Pred_polynomial[0][i], label=f'Predictions on Train Data (Polynomial) - Degree {i}')
  axs[0, 0].legend(loc='upper left')

  axs[1, 0].scatter(test_data_x_p3, test_data_y_p3, label='Test Data')
  axs[1, 0].scatter(test_data_x_p3, Y_Pred_polynomial[1][i], label=f'Predictions on Test Data (Polynomial) - Degree {i}')
  axs[1, 0].legend(loc='upper left')


  axs[0, 1].scatter(train_data_x_p3, train_data_y_p3, label='Train Data')
  axs[0, 1].scatter(train_data_x_p3, Y_Pred_gaussian[0][i], label=f'Predictions on Train Data (Gaussian) - Degree {i}')
  axs[0, 1].legend(loc='upper left')

  axs[1, 1].scatter(test_data_x_p3, test_data_y_p3, label='Test Data')
  axs[1, 1].scatter(test_data_x_p3, Y_Pred_gaussian[1][i], label=f'Predictions on Test Data (Gaussian) - Degree {i}')
  axs[1, 1].legend(loc='upper left')


  axs[0, 2].scatter(train_data_x_p3, train_data_y_p3, label='Train Data')
  axs[0, 2].scatter(train_data_x_p3, Y_Pred_sigmoidal[0][i], label=f'Predictions on Train Data (Sigmoidal) - Degree {i}')
  axs[0, 2].legend(loc='upper left')

  axs[1, 2].scatter(test_data_x_p3, test_data_y_p3, label='Test Data')
  axs[1, 2].scatter(test_data_x_p3, Y_Pred_sigmoidal[1][i], label=f'Predictions on Test Data (Sigmoidal) - Degree {i}')
  axs[1, 2].legend(loc='upper left')


  kernel_titles = ['Polynomial', 'Gaussian', 'Sigmoidal']
  for ax, kernel_title in zip(axs[0], kernel_titles):
      ax.set_title(f'Train Data - Degree {i+1} ({kernel_title})')
  for ax, kernel_title in zip(axs[1], kernel_titles):
      ax.set_title(f'Test Data - Degree {i+1} ({kernel_title})')

  plt.tight_layout()
  plt.show()

fig, axs = plt.subplots(2,1, figsize=(18, 12))
for ax in axs.flat:
    ax.set_xlabel('Polynomial Order')
    ax.set_ylabel('Error')


axs[0].plot(M_range, polynomial_tr_error, label="Training Error (Polynomial)")
axs[0].plot(M_range, gaussian_tr_error, label="Training Error (Gaussian)")
axs[0].plot(M_range, sigmoidal_tr_error, label="Training Error (Sigmoidal)")
axs[0].legend()


axs[1].plot(M_range, polynomial_te_error, label="Testing Error (Polynomial)")
axs[1].plot(M_range, gaussian_te_error, label="Testing Error (Gaussian)")
axs[1].plot(M_range, sigmoidal_te_error, label="Testing Error (Sigmoidal)")
axs[1].legend()


axs[0].set_title('Training Errors for Different Kernels')
axs[1].set_title('Testing Errors for Different Kernels')

plt.tight_layout()
plt.show()




#Plot training error vs polynomial order and testing error vs polynomial order for all the three different kernels



########################################
#Repeat the experiments by changing target function
########################################
e = np.random.normal(0, 0.1, num_points_p3)

def target(x_value, e):
  for x in x_value:
    if x in range(0, 1):
      return np.sin(2 * np.pi * x) + e
    elif x in range(1, 2):
      return ((np.abs(x - 1)) + e)
    else:
      return (np.exp(- ((x - 2)** 2) / 0.1) + e)

x_values_sin = np.random.uniform(0, 1, size = 100)
x_values_tri = np.random.uniform(1, 2, size = 100)
x_values_gauss = np.random.uniform(2, 3, size = 100)

y_values_sin = target(x_values_sin, e)
y_values_tri = target(x_values_tri, e)
y_values_gauss = target(x_values_gauss, e)

X_train_sin = x_values_sin[:50]
Y_train_sin = y_values_sin[:50]
X_test_sin = x_values_sin[50:]
Y_test_sin = y_values_sin[50:]


X_train_tri = x_values_tri[:50]
Y_train_tri = y_values_tri[:50]
X_test_tri = x_values_tri[50:]
Y_test_tri = y_values_tri[50:]


X_train_gauss = x_values_gauss[:50]
Y_train_gauss = y_values_gauss[:50]
X_test_gauss = x_values_gauss[50:]
Y_test_gauss = y_values_gauss[50:]

train_data_x_3b = np.concatenate((X_train_sin, X_train_tri, X_train_gauss), axis = None)
train_data_y_3b = np.concatenate((Y_train_sin, Y_train_tri, Y_train_gauss), axis = None)

test_data_x_3b = np.concatenate((X_test_sin, X_test_tri, X_test_gauss), axis = None)
test_data_y_3b = np.concatenate((Y_test_sin, Y_test_tri, Y_test_gauss), axis = None)



M_range=list(range(10))

polynomial_tr_error = []
polynomial_te_error = []
gaussian_tr_error = []
gaussian_te_error = []
sigmoidal_tr_error = []
sigmoidal_te_error = []
Y_Pred_polynomial = [[],[]]
Y_Pred_gaussian = [[],[]]
Y_Pred_sigmoidal = [[],[]]


for M in M_range:
    #Fit Mth order polynomial using three kernels i.e {Polynomial,Gaussian,Sigmoid}
  w_est_polynomial = KernelRegressionFit(train_data_x_3b, train_data_y_3b, 'polynomial', M, 0)
  w_est_gaussian = KernelRegressionFit(train_data_x_3b, train_data_y_3b, 'gaussian', M, 0)
  w_est_sigmoidal = KernelRegressionFit(train_data_x_3b, train_data_y_3b, 'sigmoidal', M, 0)

  trainpred_polynomial, testpred_polynomial, trainerror_polynomial, testerror_polynomial = KernelRegressionPred_Error(w_est_polynomial, train_data_x_3b, train_data_y_3b, test_data_x_3b, test_data_y_3b, 'polynomial')
  Y_Pred_polynomial[0].append(trainpred_polynomial)
  Y_Pred_polynomial[1].append(testpred_polynomial)
  polynomial_tr_error.append(trainerror_polynomial)
  polynomial_te_error.append(testerror_polynomial)


  trainpred_gaussian, testpred_gaussian, trainerror_gaussian, testerror_gaussian = KernelRegressionPred_Error(w_est_gaussian, train_data_x_3b, train_data_y_3b, test_data_x_3b, test_data_y_3b, 'gaussian')
  Y_Pred_gaussian[0].append(trainpred_gaussian)
  Y_Pred_gaussian[1].append(testpred_gaussian)
  gaussian_tr_error.append(trainerror_gaussian)
  gaussian_te_error.append(testerror_gaussian)

  trainpred_sigmoidal, testpred_sigmoidal, trainerror_sigmoidal, testerror_sigmoidal = KernelRegressionPred_Error(w_est_sigmoidal, train_data_x_3b, train_data_y_3b, test_data_x_3b, test_data_y_3b, 'sigmoidal')
  Y_Pred_sigmoidal[0].append(trainpred_sigmoidal)
  Y_Pred_sigmoidal[1].append(testpred_sigmoidal)
  sigmoidal_tr_error.append(trainerror_sigmoidal)
  sigmoidal_te_error.append(testerror_sigmoidal)



    #Predict training and testing targets using estimated w


    #Predict errors on both training and testing data using estimated w



    #Store them for plotting

for i in range(len(Y_Pred_polynomial[0])):
    fig, axs = plt.subplots(2, 3, figsize=(18, 12))

    for ax in axs.flat:
        ax.set_xlabel('X')
        ax.set_ylabel('Y')

    # Plot for Polynomial kernel
    axs[0, 0].scatter(train_data_x_3b, train_data_y_3b, label='Train Data')
    axs[0, 0].scatter(train_data_x_3b, Y_Pred_polynomial[0][i], label=f'Predictions on Train Data (Polynomial) - Degree {i}')
    axs[0, 0].legend(loc='upper left')

    axs[1, 0].scatter(test_data_x_3b, test_data_y_3b, label='Test Data')
    axs[1, 0].scatter(test_data_x_3b, Y_Pred_polynomial[1][i], label=f'Predictions on Test Data (Polynomial) - Degree {i}')
    axs[1, 0].legend(loc='upper left')

    # Plot for Gaussian kernel
    axs[0, 1].scatter(train_data_x_3b, train_data_y_3b, label='Train Data')
    axs[0, 1].scatter(train_data_x_3b, Y_Pred_gaussian[0][i], label=f'Predictions on Train Data (Gaussian) - Degree {i}')
    axs[0, 1].legend(loc='upper left')

    axs[1, 1].scatter(test_data_x_3b, test_data_y_3b, label='Test Data')
    axs[1, 1].scatter(test_data_x_3b, Y_Pred_gaussian[1][i], label=f'Predictions on Test Data (Gaussian) - Degree {i}')
    axs[1, 1].legend(loc='upper left')

    # Plot for Sigmoidal kernel
    axs[0, 2].scatter(train_data_x_3b, train_data_y_3b, label='Train Data')
    axs[0, 2].scatter(train_data_x_3b, Y_Pred_sigmoidal[0][i], label=f'Predictions on Train Data (Sigmoidal) - Degree {i}')
    axs[0, 2].legend(loc='upper left')

    axs[1, 2].scatter(test_data_x_3b, test_data_y_3b, label='Test Data')
    axs[1, 2].scatter(test_data_x_3b, Y_Pred_sigmoidal[1][i], label=f'Predictions on Test Data (Sigmoidal) - Degree {i}')
    axs[1, 2].legend(loc='upper left')

    # Add titles for each plot
    kernel_titles = ['Polynomial', 'Gaussian', 'Sigmoidal']
    for ax, kernel_title in zip(axs[0], kernel_titles):
        ax.set_title(f'Train Data - Degree {i+1} ({kernel_title})')
    for ax, kernel_title in zip(axs[1], kernel_titles):
        ax.set_title(f'Test Data - Degree {i+1} ({kernel_title})')

    plt.tight_layout()
    plt.show()

fig, axs = plt.subplots(2,1, figsize=(18, 12))
for ax in axs.flat:
    ax.set_xlabel('Polynomial Order')
    ax.set_ylabel('Error')


axs[0].plot(M_range, polynomial_tr_error, label="Training Error (Polynomial)")
axs[0].plot(M_range, gaussian_tr_error, label="Training Error (Gaussian)")
axs[0].plot(M_range, sigmoidal_tr_error, label="Training Error (Sigmoidal)")
axs[0].legend()


axs[1].plot(M_range, polynomial_te_error, label="Testing Error (Polynomial)")
axs[1].plot(M_range, gaussian_te_error, label="Testing Error (Gaussian)")
axs[1].plot(M_range, sigmoidal_te_error, label="Testing Error (Sigmoidal)")
axs[1].legend()


axs[0].set_title('Training Errors for Different Kernels')
axs[1].set_title('Testing Errors for Different Kernels')

plt.tight_layout()
plt.show()




















```

```python
#Understanding the bias-variance trade-off
########################################
#Generate 100 data sets of noisy sinusoidal data
########################################
import numpy as np
import matplotlib.pyplot as plt


def generate_data(N):
    x = np.linspace(0, 2*np.pi, N)
    y = np.sin(x) + np.random.normal(0, 0.1, N)
    return x, y


def gaussian_basis(x, mu, sigma=0.1):
    return np.exp(-0.5 * ((x - mu) / sigma)**2)

def regularized_least_squares(X, y, lmbda):
    I = np.eye(X.shape[1])
    w = np.linalg.inv(X.T @ X + lmbda * I) @ X.T @ y
    return w


def design_matrix(x, M):
    mu = np.linspace(0, 2*np.pi, M)
    Phi = np.column_stack([gaussian_basis(x, mu_i) for mu_i in mu])
    Phi = np.column_stack([np.ones(len(x)), Phi])  # Include bias term
    return Phi


x_true = np.linspace(0, 2*np.pi, 100)
y_true = np.sin(x_true)

N = 25
L = 100
M = 25
regularization_coeffs = [0.01, 1.0, 100.0]


for lmbda in regularization_coeffs:
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.title(f"Regularization Coefficient: {lmbda}")
    plt.plot(x_true, y_true, label="True Function", color='black', linestyle='--')

    all_estimated_curves = []
    for i in range(L):
        x, y = generate_data(N)
        Phi = design_matrix(x, M)
        w = regularized_least_squares(Phi, y, lmbda)
        y_pred = Phi @ w
        plt.plot(x, y_pred, color='blue', alpha=0.1)
        all_estimated_curves.append(y_pred)


    mean_curves_interpolated = [np.interp(x_true, np.linspace(0, 2*np.pi, len(curve)), curve) for curve in all_estimated_curves]
    mean_curve = np.mean(mean_curves_interpolated, axis=0)

    plt.subplot(1, 2, 2)
    plt.title(f"Mean of Estimated Curves (λ={lmbda})")
    plt.plot(x_true, y_true, label="True Function", color='black', linestyle='--')
    plt.plot(x_true, mean_curve, color='blue')
    plt.legend()
    plt.show()

# Observations
print("Observations:")
print("- As the regularization coefficient increases, the variance of the estimated curves decreases.")
print("- However, increasing regularization also leads to increased bias, as seen in the deviation of the mean curve from the true function.")
print("- For low regularization coefficients, the estimated curves exhibit high variance but low bias, resulting in overfitting.")
print("- For high regularization coefficients, the estimated curves have lower variance but higher bias, leading to underfitting.")
print("- There is a trade-off between bias and variance, and the choice of regularization coefficient determines the balance between them.")


########################################
#Use regularized least squares to estimate w
########################################




########################################
#Illustrate the concept of Bias-Variance trade off
########################################
#1. Chose three different regularization coefficeints (low,middle and high)
#2. For every regularization coefficient, produce two plots: one displaying 100 estimated curves, and the other showing the mean of the estimated curves alongside the original function.
#3. For three regularization coefficients, you should have a total of six plots, meaning two plots for each regularization.
#4. Using the six plots above, describe the bias-variance trade-off.
```

```python
#Real world problems
#Part(2): Breast Cancer Wisconsin (Prognostic) Data Set

from google.colab import drive
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt

drive.mount('/content/drive',force_remount=True)
#Download the data from the given link and keep it in the following location
datapath = "/content/drive/MyDrive/PRML_Assignments/Assignment_2/BreastCancer/wpbc.data"
data = []
with open(datapath) as fp:
    Lines = fp.readlines()
    for line in Lines:
      data.append(line.strip().split(',')[2:33])
data = np.array(data)
data = data.astype('float32')

np.random.seed(42)
np.random.shuffle(data)

def feature_scaling(X):
    mean = np.mean(X, axis=0)
    std = np.std(X, axis=0)
    X_scaled = (X - mean) / std
    return X_scaled, mean, std

def compute_cost(X, y, theta, lambda_reg):
    m = len(y)
    predictions = X.dot(theta)
    cost = (1/(2*m)) * np.sum((predictions - y)**2) + (lambda_reg/(2*m)) * np.sum(theta[1:]**2)
    return cost

def gradient_descent(X, y, theta, alpha, lambda_reg, num_iters):
    m = len(y)
    costs = []
    for i in range(num_iters):
        error = X.dot(theta) - y
        gradient = (1/m) * X.T.dot(error) + (lambda_reg/m) * np.r_[[[0]], theta[1:].reshape(-1, 1)]
        theta -= alpha * gradient
        cost = compute_cost(X, y, theta, lambda_reg)
        costs.append(cost)
    return theta, costs


X = data[:, :-1]
y = data[:, -1].reshape(-1, 1)


X_scaled, mean, std = feature_scaling(X)


X_scaled = np.hstack((np.ones((X_scaled.shape[0], 1)), X_scaled))


train_size = int(0.8 * len(X_scaled))
X_train, X_test = X_scaled[:train_size], X_scaled[train_size:]
y_train, y_test = y[:train_size], y[train_size:]


theta = np.zeros((X_train.shape[1], 1))
lambda_reg = 0.1
alpha = 0.01
num_iters = 1000


theta, costs = gradient_descent(X_train, y_train, theta, alpha, lambda_reg, num_iters)


y_pred_train = X_train.dot(theta)
y_pred_test = X_test.dot(theta)


train_rmse = np.sqrt(np.mean((y_pred_train - y_train)**2))
test_rmse = np.sqrt(np.mean((y_pred_test - y_test)**2))

print(f"Train RMSE: {train_rmse}")
print(f"Test RMSE: {test_rmse}")


plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(range(num_iters), costs)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost vs. Iterations')


plt.subplot(1, 2, 2)
plt.scatter(list(range(len(y_pred_test))), y_pred_test, color='blue', label='Predicted')
plt.scatter(list(range(len(y_test))), y_test, color='red', label='Actual')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Actual vs. Predicted(Scaled)')
plt.legend()

plt.tight_layout()
plt.show()


plt.figure(figsize=(8, 6))
plt.scatter(y_pred_test, y_pred_test - y_test, color='blue')
plt.hlines(y=0, xmin=min(y_pred_test), xmax=max(y_pred_test), colors='red', linestyle='--')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()


lambda_values = [0.01, 0.1, 1, 10]
train_rmse_values = []
test_rmse_values = []

for lambda_val in lambda_values:
    theta = np.zeros((X_train.shape[1], 1))
    theta, _ = gradient_descent(X_train, y_train, theta, alpha, lambda_val, num_iters)
    y_pred_train = X_train.dot(theta)
    y_pred_test = X_test.dot(theta)
    train_rmse_values.append(np.sqrt(np.mean((y_pred_train - y_train)**2)))
    test_rmse_values.append(np.sqrt(np.mean((y_pred_test - y_test)**2)))

plt.figure(figsize=(8, 5))
plt.plot(lambda_values, train_rmse_values, marker='o', label='Train RMSE')
plt.plot(lambda_values, test_rmse_values, marker='o', label='Test RMSE')
plt.xscale('log')
plt.xlabel('Regularization parameter (lambda)')
plt.ylabel('RMSE')
plt.title('Regularization Parameter vs. RMSE')
plt.legend()
plt.show()
```

1. **Learning Curve Analysis**:
   - The graph depicting the cost versus the number of iterations illustrates the model's learning process. A smooth decrease in cost over iterations indicates effective learning, while fluctuations or stagnation might imply issues with learning rate or model fitting.

2. **Actual vs. Predicted Scatter Plot**:
   - This scatter plot compares actual and predicted values, showcasing the model's predictive accuracy. Ideally, points should align closely to the diagonal line, indicating accurate predictions. Deviation from this line could reveal prediction errors, hinting at model bias or variance concerns.

3. **Residual Plot Examination**:
   - Analyzing the residual plot, which displays the difference between actual and predicted values against predicted values, helps diagnose model performance. Randomly scattered residuals around the zero line suggest homoscedasticity, while patterns or trends may indicate heteroscedasticity, requiring adjustments to the model.
